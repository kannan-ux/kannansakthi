```python
df=df.drop(25)
df['CPT Code']=df['CPT Code'].astype("int64")
```

```python
df['Payment Amount']=df['Payment Amount'].replace('[$]',"",regex=True).astype("float")
```

```python
df['Balance']=df['Balance'].replace('[$]',"",regex=True).astype("float")
df
```

```python
df['Denial Reason'].value_counts()
```

```python
df.info()
```

```python
df['Denial Reason'] = df['Denial Reason'].fillna('paid')
```

```python
df.to_csv(r"C:\Users\sakth\Downloads\billanalysis.csv")
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Insurance Company']=le.fit_transform(df['Insurance Company'])
```

```python
df['Physician Name']=le.fit_transform(df['Physician Name'])
```

```python
df.to_csv(r"C:\Users\sakth\Downloads\billinganalysis.csv")
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df[['CPT Code','Insurance Company','Physician Name','Payment Amount','Balance']]
```

```python
x
```

```python
y=df[['Denial Reason']]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,train_size=0.8,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
de=DecisionTreeClassifier()
de.fit(X_train,Y_train)
```

```python
ypred=de.predict(X_test)
from sklearn.metrics import classification_report
fullreport=classification_report(ypred,Y_test)
```
